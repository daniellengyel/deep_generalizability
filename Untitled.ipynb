```python
import numpy as np
```

```python
def func(x, y):
    if (x + y) > 1:
        return None
    return 1 - x**2 - y**2 - (1 - x - y)**2 #np.exp(-x) - np.exp(-y**2/(50*x))

def all_func(inp):
    out = []
    for i in range(inp.shape[1]):
        out.append(func(inp[0][i], inp[1][i]))
    return np.array(out)


X = np.linspace(0, 1, num=1000)
Y = np.linspace(0, 1, num=1000)


inp = np.array(np.meshgrid(X, Y)).reshape(2, len(X)*len(Y))

# X = np.linspace(0, 1, num=100)
# Y = np.array([[np.linspace(0, x, num=10)] for x in X])


out = all_func(inp)

out = out.reshape(len(X), len(Y))
```

```python
import plotly.graph_objects as go

xaxis = {
    'showgrid': False, # thin lines in the background
    'zeroline': False, # thick line at x=0
    'visible': False,  # numbers below
    }

fig = go.Figure(data=[go.Surface(z=out, x=X, y=Y, colorscale="twilight", 
                                 colorbar=dict(lenmode="fraction", len=0.5, x=0.7, y=0.45, tickfont=dict(size=14)),)],)

fig.update_layout(title='', autosize=False,
            
                  font=dict(
                        size=14,
                    ),
                  width=1920, height=1080,
#                     width=700, height=600,
                scene={"zaxis": 
                       dict(range=[0, 0.7], tickmode="array", tickvals = [0, 1/3., 2/3.], ticktext = ["0", "1/3", "2/3"],
                            title='1 - ' + u"\u03B7" + "\u00B2" + "(a)"),
                      "xaxis": dict(tickmode="array", tickvals = [0, 1/3., 2/3., 1], ticktext = ["0", "1/3", "2/3", "1"]), 
                      "yaxis": dict(tickmode="array", tickvals = [0, 1/3., 2/3., 1], ticktext = ["0", "1/3", "2/3", "1"])})
#                  "xaxis": dict(font=dict(size=25)), "yaxis": xaxis, "zaxis": xaxis})
# fig.update_yaxes(title='no', visible=False, showticklabels=False)


eye = fig.layout.scene.camera.eye
eye.x = 1.2 * 1.8
eye.y = 0.8 * 1.8
eye.z = 0.25

# fig.show()
```

```python
fig.write_image("./3d.pdf", "pdf")
```
